```python
import pandas as pd
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv('kyoto_hotel_comp.csv', index_col=0)
```

```python
df1 = df.sort_values('permit_date')
```

```python
df1
```

```python
def get_age(date):
    year = date[:4]
    year1 = str(year[:-1]) + '0'
    return int(year1)

def get_yaer(date):
    year = int(date[:4])
    return year
```

```python
df1['year'] = df1.apply(lambda row: get_yaer(row['permit_date']), axis=1)
df1['age'] = df1.apply(lambda row: get_age(row['permit_date']), axis=1)
df1['count'] = 1
```

```python
df1
```

```python
df1['year'].unique()
```

```python
df2 = df1.groupby(['year']).sum()
df2['count'].plot()
```

```python
df1.to_csv('kyoto_hotel_comp1.csv')
```
